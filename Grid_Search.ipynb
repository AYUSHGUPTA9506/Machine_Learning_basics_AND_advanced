```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
#how many types of species which i want to predict
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('species',axis=1)
y=df['species']
```

```python
X
```

```python
"""For cross-validation we do not need to use the train test split."""
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn=KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
#If I am getting the Model Score  of 100% accuracy, then my model is overfit.
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma= 'auto')#hyperplane ke decision boundary ke andar koi data aajata hai to usko kehte hai 'c' penalty
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
"""cross validation can reduce the over fitting of our model"""
```

```python
"""How can we use the parameters to its best? we will see in Grid search CV"""
```

```python
#now lets use the grid search cv
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svm),{
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},cv = 5,return_train_score = False)
```

```python
"""Cross-validation has its own built-in scissor hands—it automatically splits, rotates, and tests your data in cycles, so you don't have to slice it manually beforehand."""
```

```python
#We are using grid search cv  and then we are checking by applying in svm hyperparameters. How much hour model performances is measured
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
results
```

```python
#interested- which parameter of c and which parameter of kernel is giving the best results
```

```python
results[['param_C','param_kernel','mean_test_score']]
#mean_test_score == accuracy
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svm),{
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},n_iter=4,cv = 5,return_train_score = False)
```

```python
classifier_r.fit(X,y)
```

```python
results = pd.DataFrame(classifier_r.cv_results_)
```

```python
results[['param_C','param_kernel','mean_test_score']]
```
